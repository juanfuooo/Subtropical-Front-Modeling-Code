## In this demo, I will show the averaging in different dimentional order will result different averaging values when the dataset contains nan values.

```python
import numpy as np
import xarray as xr
```

```python
np.random.seed(0)
X = np.random.rand(3,3,3)
X[1,2,1] = np.nan
X[0,1,1] = np.nan
X[2,1,2] = np.nan
X = xr.DataArray(X, dims=('x','y','z'))
X
```

```python
x1 = X.mean(axis=0)
x3 = x1.mean(axis=0)
x3
```

```python
x2 = X.mean(axis=1)
x4 = x2.mean(axis=0)
x4
```

```python
x5 = X.values
x5 = x5.reshape(9,3)
x5
```

```python
x6 = np.nanmean(x5, axis=0)
x6
```
